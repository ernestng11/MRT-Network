```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
pd.set_option('mode.chained_assignment',None)
trips = pd.read_csv('mrt_trips_sampled.csv')
```

```python
trips
```

**Check data types for edges dataframe. I am looking to see if time is a datetime object or not**

```python
print(trips.dtypes)
print('Type of datetime object is',type(trips.iloc[0].origin_tm))
```

Time given in the data is a string, so we have to do some cleaning on the time columns later.

**Check for any missing values**

```python
for col in trips.columns:
    naCount = trips[col].isnull().sum()
    naCountpct = round(naCount/len(trips),5) * 100
    print('Column name:',col)
    print('Number of missing values:', naCount)
    print('Missing values percentage(%):', naCountpct)
    print('------------------------------')
```

There are no missing values in all the columns.

**Get all possible stations in our MRT network**

Store in set: *all_stations*

```python
all_stations = set(trips.origin.unique().tolist())
all_stations = all_stations.union(set(trips.destination.unique().tolist()))
print('There are {} possible stations in this network'.format(len(all_stations)))
```

**Clean the station names of any leading/trailing whitespaces**

```python
trips['destination'] = trips['destination'].apply(lambda x: x.strip())
trips['origin'] = trips['origin'].apply(lambda x: x.strip())
```

**Check for any rows where origin is the same as destination (self-loops)**

I want to remove these rows if any, since it does not make sense for a MRT station to have a train line that loops back to itself

```python
df = trips[trips['destination']!=trips['origin']]#.reset_index().drop(['index'], axis=1)
print('Size of original data: {}'.format(len(trips)))
print('Size of new data: {}'.format(len(df)))
print('{} self-loops removed'.format(len(trips)-len(df)))
```

**Parse string time into time(s) and get the travel time from origin to destination**

```python
#function to parse origin_tm and destination_tm into time in seconds
def getSeconds(time):
    h, m, s = time.split(':')
    return int(h)*3600 + int(m)*60 + int(s)
#create new column for travelling time
df['travel_time'] = df.apply(lambda x: getSeconds(x.destination_tm)-getSeconds(x.origin_tm), axis=1)
df.head()
```

```python
df.travel_time.describe()
```

```python
df[df.travel_time<61]
```

**Check if time_diff < 0. If there are such rows, remove them since it does not make sense to take negative time to travel from origin to destination**

```python
def checkNegative(time):
    return time < 0

df['neg_time'] = df['travel_time'].apply(lambda x: checkNegative(x))

print('There are {} rows where time difference < 0'.format(len(df[df['neg_time']])))
```

Since there are many rows where pairs are the same, I have to decide on whether to use mean/median/mode to get the final travel time. I rule out mode since this might be a unique value and focus on mean or median.

If the travel time distribution is skewed/not normal, I will take the median since it is more robust to outliers and skewed distributions. I look at the top 10 pairs with highest counts and plot their distributions. Alternatively, I can get the absolute difference between median and mean, if the difference is large, I will use median. Hence, a safe choice would be to use **median**.

For each destination node, we get the median for all unique trips that end at that node and take the shortest 2 paths. I assume that if we take the first 2 records, we will get the stations that are reachable from the destination from the left and right.

```python
pivot = pd.pivot_table(df, index=['destination','origin'],values=['travel_time'],aggfunc=np.median)
pivot = pivot.sort_values(by=['destination','travel_time'])
df2 = pd.DataFrame(pivot.to_records())
df2
```

Observe that the first 2 row where the 2 closest stations to Admiralty is Woodlands and Sembawang. If we refer to the actual MRT network, this is correct.

The 2 closest stations to Sembawang would be Admiralty and Yishun, which confirms our assumption.

**Take the top 2 shortest path for each destination station according to the sorted medians**

```python
grouped_df = df3.groupby('destination')
short = grouped_df.head(2).reset_index().drop(['index'], axis=1)
```

```python
short.head(10)
```

There are 153 unique destinations in our *short* df, which means every station is considered in at least one path in this dataset.

Since the mrt network should be represented as an undirected graph, I will remove the order of origin -> destination by adding the 2 stations in a list and sort them alphabetically to ensure I get a structured order of station pairs

```python
short['combined'] = short.apply(lambda x: ','.join(sorted(list([x['destination'], x['origin']]))),axis=1)
```

```python
short.sort_values('combined')
```

```python
pivot = pd.pivot_table(short, index=['combined'],
                       values=['travel_time'], aggfunc=np.mean).to_records()
comb = pd.DataFrame(pivot)
comb
```

```python
stations = comb.combined.tolist()
times = comb.travel_time.tolist()
edgelist = []
for i in zip(stations, times):
    st = i[0].split(",")
    edgelist.append((st[0],st[1],int(i[1])))
```

```python
G = nx.Graph()
G.add_weighted_edges_from(edgelist)
pos = nx.spring_layout(G)

plt.figure(figsize=(60,60))
nx.draw_networkx_nodes(G, pos, node_size=400, node_color='b')
#draw edge lines
elist = [(u, v) for (u, v, d) in G.edges(data=True)]
nx.draw_networkx_edges(G, pos, edgelist=elist, width=2)
#add in edge weights to edge lines
labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, font_size=20)
nx.draw_networkx_labels(G, pos, font_size=30, font_family="sans-serif")

plt.axis("off")
plt.show()
```

I notice that there are many cycles that involve 3 stations. We could remove by checking if such a cycle is formed when an edge is added. If a cycle is detected, we will remove the edge with longest travel time.

**Considerations: we can take rows where edge(Node i, Node j) is duplicated. This will give us confidence that the edge is strong. Then, using edges that have not been added in, we will add them in one by one, each time checking for a 3-way cycle and removing the edge with longest travel time.**

**Explore the undirected edges**

This dataframe shows us the travel time from Node 1 to Node 2 and Node2 to Node 1. To get the weighted edge(i, j), we can simply take the mean of both travel times. All these edges are confirmed to be present in the graph

```python
mean_edge = short[short.duplicated(subset='combined', keep=False)].sort_values('combined')
mean_edge.head(6)
```

**This dataframe contains directed edges which will iteratively add into the graph formed from mean_edge**

```python
out = short[~short.duplicated(subset='combined', keep=False)].sort_values('combined')
out
```

```python
pivot = pd.pivot_table(mean_edge, index=['combined'],
                       values=['travel_time'], aggfunc=np.mean).to_records()
comb = pd.DataFrame(pivot)
comb
```

```python
stations = comb.combined.tolist()
times = comb.travel_time.tolist()
undirected_edgelist = []
for i in zip(stations, times):
    st = i[0].split(",")
    undirected_edgelist.append((st[0],st[1],int(i[1])))
```

```python
len(undirected_edgelist)
```

```python
G = nx.Graph()
G.add_weighted_edges_from(undirected_edgelist)
pos = nx.spring_layout(G)

plt.figure(figsize=(60,60))
nx.draw_networkx_nodes(G, pos, node_size=400, node_color='r')
#draw edge lines
elist = [(u, v) for (u, v, d) in G.edges(data=True)]
nx.draw_networkx_edges(G, pos, edgelist=elist, width=2)
#add in edge weights to edge lines
labels = nx.get_edge_attributes(G,'weight')
#nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, font_size=25)
nx.draw_networkx_labels(G, pos, font_size=30, font_family="sans-serif")

plt.axis("off")
plt.show()
```

**We have many connected components scattered around. Now we iteratively add in the directed edges as undirected edges while checking that we do not form a 3-way cycle**

```python
#function to retrieve edge weight between 2 nodes
def checkEdgeW(edgepair, G=G): #input tuple (node i, node j) and graph G
    try:
        res = G.get_edge_data(edgepair[0], edgepair[1])
        return res['weight']
    except:
        return False
    
#function to remove longest edge from a cycle with 3 nodes
def fixCycle3(lst):
    #n1,n2. n2,n3. n1,n3
    n1, n2, n3 = lst[0], lst[1], lst[2]
    maxW = 0
    perms = [(n1, n2), (n2,n3), (n1,n3)]
    i = 0
    rm = 0
    for p in perms:
        if checkEdgeW(p) > maxW:
            maxW = checkEdgeW(p)
            rm = i
        i+=1
    print('Did not add', perms[rm])
    G.remove_edge(perms[rm][0], perms[rm][1])

#function to add edge then check for cycles. If any, remove longest edge 
def checkAdd(tup, G):
    n1, n2, w = tup[0], tup[1], tup[2]
    #check if n1 and n2 exists in G        
    G.add_edge(n1,n2,weight=w)
    try:
        cyc = list(filter(lambda x: len(x) == 3, list(nx.cycle_basis(G))))
        for c in cyc:
            fixCycle3(c)
    except:
        return
```

```python
len(G.edges)
```

**Prepare edge list for the edges not added in yet**

```python
stations2 = out.combined.tolist()
times2 = out.travel_time.tolist()
undirected_edgelist2 = []
for i in zip(stations2, times2):
    st = i[0].split(",")
    undirected_edgelist2.append((st[0],st[1],int(i[1])))
```

```python
for de in undirected_edgelist2:
    checkAdd(de, G)
```

**This graph contains lesser 3-way cycles than the first graph**

Midway Findings: Taking the mean/median and finding the 2 shortest path relative to the origin, might not always give us the adjacent stations to the origin. This approach will not give us a confident result that all stations will be connected in some way.
